## Program of Thoughts Rationale

```python
import json
import random

with open('theoremqa_test.json') as f:
    test_set = json.load(f)
mapping = {entry['id']: entry for entry in test_set}
                 
with open('outputs/GPT4_PoT_s0_e-1_05_13_10_18.jsonl.corrected') as f:
    output = json.load(f)
random.shuffle(output)

for entry in output[:10]:
    if entry['answer_type'] not in ['bool', 'option'] and entry['rationale']:
        print(entry['id'])
        print('Question:', entry['question'])
        print('Theorem:', mapping[entry['id']]['theorem'])
        print('Program:')
        print(entry['rationale'])
        print('prediction:', entry['prediction'])
        print('answer:', entry['answer'])
        print('----------------------')
```
